```python
import pandas as pd
```

```python
df=pd.read_csv('train.csv')
df_t=pd.read_csv('test.csv')
```

```python
from pycaret.regression import *
```

```python
exp_name=setup(data=df,target = 'SalePrice', 
                   normalize = True,
                   )
```

```python
best_model=compare_models()
```

```python
# Creating models for the best estimators 
Passive_Aggressive_Regressor = create_model('par')
Bayesian_Ridge = create_model('br')
Lasso_Least_Angle_Regression = create_model('llar')
Elast = create_model('en')
Huber_Regressor=create_model('huber')
```

### Modellerin tuningini yapıyoruz. r2 scrone yükseltmesi için 0,86 den 0,95 lere çıktı

```python
Passive_Aggressive_Regressor = tune_model(Passive_Aggressive_Regressor)
Bayesian_Ridge = tune_model(Bayesian_Ridge)
Lasso_Least_Angle_Regression = tune_model(Lasso_Least_Angle_Regression)
Elast = tune_model(Elast)
Huber_Regressor = tune_model(Huber_Regressor)
```

```python
# Blending models- Burada modellerin arasında oylama yapıyor
blender = blend_models(estimator_list = [Passive_Aggressive_Regressor, Bayesian_Ridge,Lasso_Least_Angle_Regression,Elast,Huber_Regressor])
```

```python
blender
```

```python
# Finaliszing model for predictions - Tununing i yapılmış modeller ile test datasında tahmin yapacağız.

model = finalize_model(blender)

predictions = predict_model(model, data = df_t)
```
